```python
%matplotlib widget
import matplotlib.pyplot as plt
import numpy as np
from numpy.random import default_rng
np.set_printoptions(linewidth=1000, precision=6)

rng = default_rng()
vals = rng.standard_normal(10)
more_vals = rng.standard_normal(10)
print(vals, "\n", more_vals)
```

```python
mean = [0, 0]
cov = [[1, 0], [0, 100]]
x, y = rng.multivariate_normal(mean, cov, 5000).T
plt.plot(x, y, 'x')
plt.axis('equal')
plt.show()
```
